# 5. Move the dataset into a Training and Testing Partition

```python
import shutil
import os
import fnmatch
import math

dir=r'C:\Users\Varshaa\OneDrive\Desktop\CI_AI\Tensorflow\workspace\images\collectedimages'
train_path=r'C:\Users\Varshaa\OneDrive\Desktop\CI_AI\Tensorflow\workspace\images\train'
test_path=r'C:\Users\Varshaa\OneDrive\Desktop\CI_AI\Tensorflow\workspace\images\test'

test=0.1


for folder in os.listdir(dir):
    a=os.path.join(dir,folder)
    count = len(fnmatch.filter(os.listdir(a), '*.*'))
    train_files=2*(math.floor((1-test)*(count/2)))
    proc=0
    for files in os.listdir(a):
        if(proc<train_files):
            file_source=os.path.join(a,files)
            proc=proc+1
            shutil.move(file_source,train_path)

for folder in os.listdir(dir):
    a=os.path.join(dir,folder)
    for files in os.listdir(a):
        file_source=os.path.join(a,files)
        shutil.move(file_source,test_path)
```
